```python
import cv2
import numpy as np
import matplotlib.pyplot as plt 
```

# Negation of an image
- negated_pixel_value = 255 - original_pixel_value


```python
img = cv2.imread("Spatial Domain/img.jpg")
negative_img = np.zeros(img.shape , dtype=int)
m , n , k =  img.shape
for k in range(3):
    for i in range(0 , m-1):
        for j in range(0 , n-1):
            negative_img[i,j ,k] = 255 - img[i,j , k]
```

```python
plt.figure(figsize=(13,5))
plt.subplot(121)
plt.title("Original")
plt.imshow(img)
plt.subplot(122)
plt.title("Negative")
plt.imshow(negative_img)
```

# Thresholding of an image

```python
Threshold , Threshold_img = cv2.threshold(img , 127 , 255 , cv2.THRESH_BINARY)
plt.figure(figsize=(13,5))
plt.subplot(121)
plt.title("Original")
plt.imshow(img)
plt.subplot(122)
plt.title("Threshold with 127")
plt.imshow(Threshold_img)
```

```python
t , Threshold_img =  cv2.threshold(img,127,255,cv2.THRESH_BINARY_INV)
plt.figure(figsize=(13,5))
plt.subplot(121)
plt.title("Original")
plt.imshow(img)
plt.subplot(122)
plt.title("Threshold with 127")
plt.imshow(Threshold_img)
```

# Contrast Stretching of an image
- NewPixelValue = (OriginalPixelValue - OldMin) * (NewMax - NewMin) / (OldMax - OldMin) + NewMin


```python
Constrast_img = cv2.imread("Spatial Domain/img.jpg")
m , n , k =  img.shape
for k in range(3):
    for i in range(0 , m-1):
        for j in range(0 , n-1):
            Constrast_img[i,j ,k] = (img[i,j , k] - img.min())*(Constrast_img.max()
                -Constrast_img.min()) / (img.max()-img.min()) + Constrast_img.max()
```

```python
plt.figure(figsize=(13,5))
plt.subplot(121)
plt.title("Original")
plt.imshow(img)
plt.subplot(122)
plt.title("Constrast")
plt.imshow(Constrast_img)
```

# Image Transformation

```python
c = 255/(np.log(1 + np.max(img))) 
log_transformed = c * np.log(1 + img) 
log_transformed = np.array(log_transformed, dtype = np.uint8) 
```

```python
plt.figure(figsize=(13,5))
plt.subplot(121)
plt.title("Original")
plt.imshow(img)
plt.subplot(122)
plt.title("log_transformed")
plt.imshow(log_transformed)
```
